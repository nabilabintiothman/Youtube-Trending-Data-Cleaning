# Data cleaning project

This is a simple data cleaning project which I did when I started learning Python.

Towards the end of this notebook, I also did some simple data extraction. 

```python
#Import libraries

import numpy as np 
import pandas as pd 

# Input data files are available in the read-only "../input/" directory
# For example, running this (by clicking run or pressing Shift+Enter) will list all files under the input directory
import os
for dirname, _, filenames in os.walk('/kaggle/input'):
    for filename in filenames:
        print(os.path.join(dirname, filename))
```

```python
#Load the Data into Pandas Dataframe
youtube = pd.read_csv("/kaggle/input/youtube-new/GBvideos.csv")
```

```python
#view the first five rows of the dataframe
youtube.head()
```

```python
#number of rows and features in the data and each of the features types
youtube.info()
```

This youtube dataset has 38916 rows and 16 columns i.e features.

```python
#number of missing value in youtube dataset
missing_values_count = youtube.isnull().sum()
missing_values_count

#only description column has missing values
```

```python
#percentage of missing values 

# how many total missing values do we have?
total_cells = np.product(youtube.shape)
total_missing = missing_values_count.sum()

# percent of data that is missing
(total_missing/total_cells) * 100
```

```python
#one method to find the index of null values to check if we have correctly replace the missing values
#can also just check df.isnull().sum()
youtube.loc[pd.isna(youtube["description"]), :].index

#another method
#indices = list(np.where(youtube['description'].isnull())[0])
```

Only the description column has null values.
We can either drop the whole column if the null values percentage is high or drop only the rows where there is null values.

Another technique to handle missing values is to replace the missing values or to impute them for numerical data.

For this project, I will replace the missing values with "no description available"

```python
# replace the null values with "no description available for this video"
youtube = youtube.fillna("no description available for this video")

#check at index 91
youtube.loc[91]
```

```python
#youtube.isnull().sum() to check missing values after replaced
#can also use youtube.info()
youtube.info()
```

```python
#find complete duplicates

column_names = ['title','trending_date']
duplicates2 = youtube.duplicated(subset = column_names, keep=False)
youtube[duplicates2].sort_values(by='title')

#there are 348 rows with duplicates (same video and same trending date and other columns also same)
```

```python
youtube['title'].loc[34137]
```

```python
#drop duplicates of column title and trending date and keep the first entry only
youtube = youtube.drop_duplicates(['title','trending_date'],keep= 'first')

#check if we have successfully remove duplicates. i.e should not have same trending date for each title
youtube[youtube['title'] == '13 Reasons Why: Season 2 | Official Trailer [HD] | Netflix' ]
```

```python
#There are columns with inaccurate data types and we need to fix them
#change trending_date to datetime object
#change publish_time to datetime object

youtube['trending_date'] = pd.to_datetime(youtube['trending_date'], format="%y.%d.%m")
youtube['publish_time'] = pd.to_datetime(youtube['publish_time'])
youtube.info()
```

```python
#statistical summary

youtube.describe()
```

```python
#save the dataset after processed
youtube.to_csv('youtube_edited.csv', index=False)
```

- The data cleaning has been completed.
- I have replace missing values, remove duplicates and make sure data types are correct.

```python
# maximum views, likes and dislikes for each category

youtube.groupby('category_id').agg({'views':'max','likes':'max','dislikes':'max'})
```
